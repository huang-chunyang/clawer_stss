```python
import pandas as pd
import numpy as np
import re
```

```python
event = "1500"
```

读取比赛数据

```python
race = pd.read_csv("data/" + event + "m.csv")
```

```python
x, y = race.shape
```

```python
race_new_names = []
```

```python
for i in range(x):
    race_new_names.append(race["Name"][i].title())
```

```python
race["Name"]= race_new_names
```

```python
race
```

运动员数据

```python
athletes = pd.read_csv("data/new_athletes.csv")[["Name", "Gender", "Age"]]
```

```python
x, y = athletes.shape
athletes_gender_list = []
for index in range(x):
    athletes_gender_list.append(athletes["Gender"][index][0])
athletes["Gender"] = athletes_gender_list
```

```python
old_athletes = pd.read_csv("data/athletes.csv")
rows, columns = old_athletes.shape
old_athletes
```

```python
age_list = []
name_list = []
year = 2022
for i in range(rows):
    birthyear = re.match(".+(\d{4})", old_athletes["Birthdate"][i]).group(1)
    age_list.append(year - int(birthyear))
    name_list.append(old_athletes["Name"][i].title())
old_athletes["Age"] = age_list
old_athletes["Name"] = name_list
```

```python
old_athletes = old_athletes[["Name", "Gender", "Age"]]
```

```python
old_athletes
```

```python
all_athletes = pd.concat([athletes, old_athletes]).drop_duplicates()
```

```python
res = pd.merge(race, all_athletes, how="left", on="Name")
```

```python
all_athletes.to_csv("res_data/all_athletes.csv")
```

```python
res.to_csv("res_data/race_" + event + "m.csv")
```
